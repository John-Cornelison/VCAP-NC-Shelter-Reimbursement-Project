```python
#functions

def cpi_year_inflate(value, start_year):
    start_year_df = cpi_index[cpi_index["Year"] == start_year]
    current_year_df = cpi_index[cpi_index["Year"] == 2023]
    interest_rate = current_year_df["Nov"].values[0]/start_year_df["Dec"].values[0]
    new_value = round(value*interest_rate, 2)
    
    return(new_value)


def cpi_south_year_inflate(value, start_year):
    start_year_df = cpi_south_index[cpi_south_index["Year"] == start_year]
    current_year_df = cpi_south_index[cpi_south_index["Year"] == 2023]
    interest_rate = current_year_df["Nov"].values[0]/start_year_df["Dec"].values[0]
    new_value = round(value*interest_rate, 2)
    
    return(new_value)
```

```python
species_table = pd.read_csv("species_table.csv")
reimbursemnets_table = pd.read_csv("reimbursements_table.csv")
expenses_table = pd.read_csv("expenses_table.csv")
cpi_index = pd.read_excel("Jan_1913toNov_2023.xlsx")
cpi_south_index = pd.read_excel("SouthJan_1987toNov_2023.xlsx")
```

```python
cpi_year_inflate(200, 2010)
```

```python
cpi_south_year_inflate(155.69, 2010)
```

```python
inf_adj_operating_expense_list = []
inf_adj_cost_per_animal = []

for val, year in zip(expenses_table["Operating_Expense"], expenses_table["Year"]):
    inf_adj_operating_expense_list.append(cpi_year_inflate(val, year))
    
for val, year in zip(expenses_table["Cost_Per_Animal"], expenses_table["Year"]):
    inf_adj_cost_per_animal.append(cpi_year_inflate(val, year))
```

```python
south_inf_adj_operating_expense_list = []
south_inf_adj_cost_per_animal = []

for val, year in zip(expenses_table["Operating_Expense"], expenses_table["Year"]):
    south_inf_adj_operating_expense_list.append(cpi_south_year_inflate(val, year))
    
for val, year in zip(expenses_table["Cost_Per_Animal"], expenses_table["Year"]):
    south_inf_adj_cost_per_animal.append(cpi_south_year_inflate(val, year))
```

```python
expenses_table["Operating_Expense_inf_adj"] = inf_adj_operating_expense_list
expenses_table["Cost_Per_Animal_inf_adj"] = inf_adj_cost_per_animal
expenses_table["Operating_Expense_south_inf_adj"] = south_inf_adj_operating_expense_list
expenses_table["Cost_Per_Animal_south_inf_adj"] = south_inf_adj_cost_per_animal
```
